```python
import pandas as pd
import numpy as np
import random
import csv

nos = [0,1,2,3,4]
frequency = [0,0,0,0,0]
stop_frequency = [50,150,300,300,200]
range_start = [70,75,75,80,85]
range_end = [78,83,85,90,93]

no_gender=[0,1]

csv_data = [['Gender','Designation','Accuracy']]

count=1
while count <=1000:
    gender= random.choice(no_gender)
    random_no = random.choice(nos)
    frequency[random_no]+=1
    if(frequency[random_no]>=stop_frequency[random_no]):
        nos.remove(random_no)
    csv_data.append([gender,random_no,random.choice(range(range_start[random_no],range_end[random_no]+1))])
    count+=1
print(frequency)

rows = [random_no]
with open('model_2.csv', 'w',newline='') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerows(csv_data)
csvFile.close()

#Method to set threshold for predicted value
def custom_accuracy(y_test,y_pred,thresold):
    right = 0
    l = len(y_pred)
    for i in range(0,l):
        if(abs(y_pred[i]-y_test[i]) <= thresold):
            right += 1
    return ((right/l)*100)


# Importing the dataset
dataset=pd.read_csv("C:\\Users\\Abhi\\model_2.csv")
X = dataset.iloc[:,[0,1]].values
y = dataset.iloc[:, 2].values


# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)


# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


# Training the dataset
from sklearn.tree import DecisionTreeClassifier
reg=DecisionTreeClassifier()
reg.fit(X_train,y_train)


# Testing the dataset on trained model
y_pred = reg.predict(X_test)
score = reg.score(X_test,y_test)*100
#print("R square value:" , score)
#print("Custom accuracy:" , custom_accuracy(y_test,y_pred,10))


list1=[]
for i in range(0,2):
    for j in range(0,5):
        list2=[reg.predict(sc.transform(np.array([[i,j]])))]
        #print(list2[0])
        list1.append(list2)
        
print("Predicted Accuracy for a Group:",max(list1)[0])
```
